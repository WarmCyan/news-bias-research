```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open("../data/output/lstm_run_1/selection_biased_mbfc_13_lstm_1_15_500_128_0.001.json", 'r') as infile:
    results = json.load(infile)
results
```

```python
sns.set()
fig, axs = plt.subplots(2, figsize=(6, 8))

x = list(range(0, len(results["history"]["binary_accuracy"])))


fig.suptitle("The Filename")
axs[0].plot(x, results["history"]["binary_accuracy"], label="Training")
axs[0].plot(x, results["history"]["val_binary_accuracy"], label="Validation")
axs[1].plot(x, results["history"]["loss"], label="Training")
axs[1].plot(x, results["history"]["val_loss"], label="Validation")

axs[0].set_ylabel("Accuracy")
axs[1].set_ylabel("Loss")
axs[1].set_xlabel("Epoch")

axs[0].legend()
axs[1].legend()

#fig.savefig("test.png")

#axs[0].grid()
#axs[1].grid()
```
