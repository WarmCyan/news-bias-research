```python
def find_source_sentic_average(df, sentics, source):
    indices = list(df[df.source == source].index)
    sentics_df = pd.DataFrame(sentics)
    averaged = sentics_df.iloc[indices,0:6].mean(axis=0)
    
    data = {
        "polarity":averaged.iloc[0],
        "intensity":averaged.iloc[1], 
        "pleasantness":averaged.iloc[2], 
        "attention":averaged.iloc[3], 
        "sensitivity":averaged.iloc[4],
        "aptitude":averaged.iloc[5]
    }
    return data
```

```python
rows = []
for source in list(set(df.source)):
    data = find_source_sentic_average(df, sentics, source)
    data["source"] = source
    rows.append(data)
    
#pd.DataFrame(rows)
```

```python
rows = []

for i in range(0, 10):
    df = pd.read_pickle(f"../data/cache/binary_fold_1000_{i}")
    sentics = np.array(pd.read_pickle(f"../data/cache/biased_13_1000_fold{i}_sentic/avg_std.pkl.pkl"))

    for source in list(set(df.source)):
        if source == "Spiegel":
            continue
        data = find_source_sentic_average(df, sentics, source)
        data["source"] = source
        data["biased"] = df[df.source == source].iloc[0].biased
        rows.append(data)
        
sentic_agg_df = pd.DataFrame(rows)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(sentic_agg_df)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X = sentic_agg_df[['polarity', 'intensity', 'pleasantness', 'attention', 'sensitivity', 'aptitude']]
y = sentic_agg_df["biased"]

#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
model = LogisticRegression(random_state=13).fit(X, y)
```
